# Types of charts


There are many types of plots and charts used in data visualization, each suited for different kinds of data and analysis. Here’s a categorized list of common ones:

**1. Comparison Charts**
Bar Chart (Vertical/Horizontal) – Compares categories.

**Column Chart** – Like a bar chart but vertical.
**
**Grouped Bar/Column Chart**** – Compares sub-categories within larger groups.

**Stacked Bar/Column Chart** – Shows part-to-whole relationships in categories.

**Dot Plot** – Displays individual data points for comparison. 

**
**2. Trend & Time-Based Charts**** 
 

**Line Chart **– Shows trends over time.
**
**Area Chart**** – Like a line chart but filled under the line.

**Stacked Area Chart** – Shows cumulative trends over time.

**Step Chart** – Shows changes at specific intervals. 

**
**3. Distribution Charts**** 


**Histogram** – Shows frequency distribution of numerical data.

**Box Plot** (Box-and-Whisker Plot) – Displays data distribution and outliers.

**Violin Plot** – Combines box plot and density estimation.

**Density Plot** – Smooth histogram showing data distribution.

**4. Part-to-Whole Charts** 

**Pie Chart** – Shows proportions of a whole (best for few categories).

**Donut Chart** – A pie chart with a hole in the center.

**Treemap** – Displays hierarchical part-to-whole relationships.
**
**Sunburst Chart**** – Radial hierarchical part-to-whole visualization.
**
**5. Correlation & Relationship Charts****
****Scatter Plot** **– Shows relationship between two variables.

**Bubble Chart** – Like a scatter plot with a third variable (size).

**Heatmap** – Uses color to show correlation in a matrix.

**Parallel Coordinates Plot** – Compares multiple variables.

**6. Geospatial & Map-Based Charts**
**Choropleth Map** – Colored regions based on data values.

**Bubble Map** – Overlays bubbles on geographic locations.

****Cartogram ****– Distorts map regions based on a variable.

**Flow Map** – Shows movement between locations.

**7. Hierarchical & Network Charts**
**Tree Diagram** – Shows parent-child relationships.

**Sankey Diagram** – Visualizes flow between stages.

**Network Graph** – Displays connections between nodes.

**8. Statistical & Advanced Charts**
**Q-Q Plot** – Compares probability distributions.

**Pareto Chart** – Combines bars and a line (80/20 rule).

****Control Chart ****– Monitors process stability over time.

**9. Specialized & Modern Charts**
**Gantt Chart** – Project timeline visualization.

**Radar Chart (Spider Chart)** – Compares multivariate data.

**Waterfall Chart** – Shows incremental changes.

****Candlestick Chart** **– Financial price movements.

```python

# Sample data for demonstration
np.random.seed(0)
data = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
df = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Value': np.random.randint(10, 100, 100),
    'Value2': np.random.randint(10, 100, 100),
    'Time': np.arange(100),
    'Group': np.random.choice(['G1', 'G2'], 100)
})
```

```python
# 1. Comparison Charts

# Bar Chart
plt.figure(figsize=(4,2))
plt.bar(categories, values)
plt.title("Bar Chart: Compares categories")
plt.show()

# Grouped Bar Chart
grouped = df.groupby(['Category', 'Group']).size().unstack()
grouped.plot(kind='bar')
plt.title("Grouped Bar Chart: Sub-categories within groups")
plt.show()

# Stacked Bar Chart
grouped.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart: Part-to-whole relationships")
plt.show()

# Dot Plot
plt.figure(figsize=(4,2))
plt.plot(categories, values, 'o')
plt.title("Dot Plot: Individual data points")
plt.show()
```

# # 2. Trend & Time-Based Charts

```python
# 2. Trend & Time-Based Charts

# Line Chart
plt.figure(figsize=(4,2))
plt.plot(df['Time'], df['Value'])
plt.title("Line Chart: Trends over time")
plt.show()

# Area Chart
plt.fill_between(df['Time'], df['Value'], alpha=0.5)
plt.title("Area Chart: Filled under the line")
plt.show()

# Stacked Area Chart
df_area = pd.DataFrame({
    'A': np.random.randint(1, 10, 20),
    'B': np.random.randint(1, 10, 20)
})
df_area.plot.area()
plt.title("Stacked Area Chart: Cumulative trends")
plt.show()

# Step Chart
plt.step(df['Time'], df['Value'])
plt.title("Step Chart: Changes at intervals")
plt.show()
```

**which plot is used when?** 


Choosing the right plot depends on the type of data you have and the message you want to convey. Here’s a quick guide to help you decide which plot to use in different scenarios:

**1. Comparing Categories** 


**Bar Char**t → Compare quantities across different categories (e.g., sales by product).

**Column Chart** → Similar to a bar chart but vertical (better for fewer categories).
**
**Grouped Bar/Column Chart**** → Compare sub-categories within larger groups (e.g., sales by product and region).

**Dot Plot** → Compare individual values clearly (good for small datasets).

**Example:**

**"Compare revenue across different stores.**" → Bar Chart
**
**"Compare sales of different products in multiple regions.****" → Grouped Bar Chart
**
**2. Showing Trends Over Time**** 


**Line Chart** → Best for continuous trends (e.g., stock prices over months).

**Area Chart** → Emphasize magnitude of change over time.

**Stacked Area Chart** → Show cumulative trends (e.g., market share over time).

**Ex**ampl**e:
**
"**Show monthly website traffic growth**." → Line Chart

"Display total sales with breakdown by product category." → Stacked Area Chart
******
3.**** Distribution of Dat****a****** 


**Histogram** → Show frequency distribution of a single variable (e.g., age distribution).

**Box Plot** → Compare distributions and identify outliers (e.g., salaries across departments).

**Violin Plot** → Show distribution density and summary stats (like a smoothed box plot).

**Example:**

**"Analyze exam score distribution**." → Histogram

"**Compare income distribution across citie**s." → Box Plot

**1. Part-to-Whole Relationships** 

**Pie Chart** → Simple proportions (use only for ≤5 categories).

**Donut Char**t → Like a pie chart but cleaner (e.g., market share).

**Treemap** → Show hierarchical proportions (e.g., budget allocation).

Example:
******
"Show percentage of sales by product type******." → Pie/Donut Chart

**"Visualize company expenses by department**." → Treemap

**5. Correlation & Relationships** 
**
**Scatter Plot**** → Check relationship between two variables (e.g., height vs. weight).

**Bubble Chart** → Add a third dimension (size) to a scatter plot (e.g., GDP vs. population).

H**eatmap** → Show correlation matrix (e.g., feature correlations in data science).

Example:
****
"Does study time affect exam scores**?"** → Scatter Plot

"Compare countries by GDP, population, and life expectancy." → Bubble Chart
****
6.**** Geospatial Data****

Choropleth Map → Color regions by value (e.g., election results by state).

Bubble Map → Show values as bubbles on a map (e.g., population by city).

Example:

"Visualize unemployment rates by country." → Choropleth Map

"Show store locations and their sales volumes." → Bubble Map
**
7. Hierarchical/Network Data**  
8. 
Tree Diagram → Show parent-child relationships (e.g., organizational structure).

Sankey Diagram → Show flows (e.g., customer journey).

Network Graph → Display connections (e.g., social networks).

Example:

"Trace the path of website visitors." → Sankey Diagram

"Show relationships in a social network." → Network Graph

8. Statistical & Special Cases
Control Chart → Monitor process stability (e.g., manufacturing quality).

Waterfall Chart → Show incremental changes (e.g., profit & loss).

Gantt Chart → Track project timelines.

Example:

"Explain quarterly financial changes." → Waterfall Chart

"Track project task deadlines." → Gantt Chart
