# Animation plot

````python
# Scatter plot on gapminder dataset: lifeExp vs gdpPercap, colored by continent, sized by population
fig_gapminder_scatter = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    hover_name="country",
    log_x=True,
    title="Gapminder: Life Expectancy vs GDP per Capita"
)
fig_gapminder_scatter.show()
````

````python
# Create a choropleth mapbox using the election data and geojson
fig_election_mapbox = px.choropleth_mapbox(
    election_df,
    geojson=election_geojson,
    locations="district",
    featureidkey="properties.district",
    color="Bergeron",
    mapbox_style="carto-positron",
    zoom=9,
    center={"lat": 45.5517, "lon": -73.7073},
    opacity=0.5,
    title="Election Results: Bergeron Votes by District"
)
fig_election_mapbox.show()
````

````python
# Create a new column to determine the winner in each district
election_df["winner"] = election_df[["Bergeron", "Coderre", "Joly"]].idxmax(axis=1)

# Assign a color to each winner
winner_color_map = {"Bergeron": "blue", "Coderre": "red", "Joly": "green"}
election_df["winner_color"] = election_df["winner"].map(winner_color_map)

# Plot the choropleth mapbox colored by the winner
fig_election_mapbox = px.choropleth_mapbox(
    election_df,
    geojson=election_geojson,
    locations="district",
    featureidkey="properties.district",
    color="winner",
    color_discrete_map=winner_color_map,
    mapbox_style="carto-positron",
    zoom=9,
    center={"lat": 45.5517, "lon": -73.7073},
    opacity=0.5,
    title="Election Results: District Winner"
)
fig_election_mapbox.show()
````

# save in Gif file

````python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings("ignore")
````

````python
df = px.data.gapminder() # load the gapminder dataset

# view the first few rows of the dataset
df.head()
````

````python
fig = px.sunburst(df, path=['continent', 'country','year'], 
                  values='pop',color='continent')
fig.show()
````

````python
import plotly.express as px
import pandas as pd
import numpy as np
import io
import PIL


fig = px.scatter(df, x= "gdpPercap",
            y = "lifeExp",
            size= "pop", color= "continent",
            animation_frame='year', animation_group="country",
            log_x=True, size_max=55, range_x=[100,100000], range_y=[5,100])

fig.show()

# generate images for each step in animation
frames = []
for s, fr in enumerate(fig.frames):
    # set main traces to appropriate traces within plotly frame
    fig.update(data=fr.data)
    # move slider to correct place
    fig.layout.sliders[0].update(active=s)
    # generate image of current state
    frames.append(PIL.Image.open(io.BytesIO(fig.to_image(format="png", scale=3))))
    
# create animated GIF
frames[0].save(
        "gapminder.gif",
        save_all=True,
        append_images=frames[1:],
        optimize=True,
        duration=500,
        loop=0,
        dither=None  # Turn off dithering
    )
````
